```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tls
import warnings
from collections import Counter
from sklearn.feature_selection import mutual_info_classif

train=pd.read_csv("train.csv")
test=pd.read_csv("test.csv")
train_imp=train
train.head()
#test.head()
```

```python
train_imp.drop(['ps_ind_10_bin','ps_ind_11_bin','ps_ind_12_bin'],1,inplace=True)
train_imp.drop(['ps_ind_13_bin','ps_car_03_cat','ps_car_04_cat','ps_car_05_cat','ps_reg_03'],1,inplace=True)
train_imp.drop(['ps_calc_15_bin','ps_calc_16_bin','ps_calc_17_bin','ps_calc_18_bin','ps_calc_19_bin','ps_calc_20_bin'],1,inplace=True)
```

```python
#train-test-split

from sklearn.model_selection import train_test_split

X=train_imp
y=train['target']

train_X, test_X, train_y, test_y = train_test_split(X, y, 
                                                    train_size=0.8,
                                                    test_size=0.2,
                                                    random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
gnb = GaussianNB() #object created for Naive Bayes Class
y_pred = gnb.fit(train_X,train_y).predict(train_X)# training the model using trainind data set
y_pred_test = gnb.fit(test_X,test_y).predict(test_X) # tarining and predicting using test data set

print("Number of mislabeled points (train) out of a total %d points : %d" % (train_X.shape[0],(train_y != y_pred).sum()))
print("Number of mislabeled points (for test case) out of a total %d points : %d" % (test_X.shape[0],(test_y != y_pred_test).sum()))

gnb_train_y_acc=metrics.accuracy_score(train_y,y_pred)
print("train accuracy: %f" % (gnb_train_y_acc))

gnb_test_y_acc=metrics.accuracy_score(test_y,y_pred_test)
print("test accuracy: %f"%(gnb_test_y_acc))
```
